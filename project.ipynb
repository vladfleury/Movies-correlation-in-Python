### First let's import all the libraries and the dataset

```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # adjusts plot configurations that we will create


# import the data
df = pd.read_csv("https://raw.githubusercontent.com/vladfleury/Movies-correlation-in-Python/main/movies.csv")
```

# 1. Data Cleaning
### 1.1 Let's take a look at the data

```python
df.head()
```

### 1.2 Now let's see if there's any missing data. 

Let's show the percentage of missing values in each column

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

- As we can see - quite a lot is missing. 

### 1.3 Let's remove the missing values with panda's dropna() function

```python
df = df.dropna()
```

### 1.4 Now We'll check the datatypes of each column

```python
df.dtypes 
```

### 1.5 Let's remove the ".0" at the end of gross and budget columns by changing the data type to int.

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#pd.set_option('display.max_rows', 30)
```

### 1.6 Let's drop any duplicates if we have any

```python
df.drop_duplicates().head()
```

### 1.7 Let's check the most grossing movies

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

# 2. Finding correlations
### Hypothesis:
- Budget will have high correlation with gross
- Company will have high correlation with gross

### 2.1 Scatter plot with budget vs gross revenue

```python
plt.scatter(x = df['budget'], y = df['gross'], color = "green")

plt.title("Budget vs. Gross Earnings")

plt.xlabel("Budget")

plt.ylabel("Gross earnings")

plt.show()
```

```python
df.head()
```

### 2.2 Budget vs Gross using Seaborn

```python
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "green"}, line_kws = {"color": "blue"})
```

- We can see some positivie correlation
### 2.3 Let's look at the correlation

```python
df.corr(method = "pearson")
```

- High correlation between budget and gross revenue :D 
- We were right

### 2.4 Let's visualize our correlation

```python
correlation_matrix = df.corr(method = "pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for numeric features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

- Not all columns are displayed, because some of them are objects - we need to transform them into categorical values.
### 2.5 Let's look at all columns now

#### Transforming objects into categories

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

#### Looking at all columns

```python
correlation_matrix = df_numerized.corr(method = "pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for numeric features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
cor_mat = df_numerized.corr()
cor_pairs = cor_mat.unstack()

print(cor_pairs)
```

```python
sorted_pairs = cor_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > .5)]
high_corr
```
